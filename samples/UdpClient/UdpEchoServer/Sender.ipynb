```
using System;
using System.Net;
using System.Net.Sockets;
using System.Text;


UdpClient u = new UdpClient("192.168.0.1",7);
u.EnableBroadcast  =true;
u.Client.ReceiveTimeout = 5000; // 5sec
byte[] message = Encoding.UTF8.GetBytes($"Message sent at {DateTime.UtcNow:F}");
u.Send(message);

IPEndPoint ipEndpoint;
Console.WriteLine("Waiting for answer");
var result = u.Receive(ref ipEndpoint);
Console.WriteLine($"Got answer: {Encoding.UTF8.GetString(result)}");
```
